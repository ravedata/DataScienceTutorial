# Validation Set Approach

10 different train-test splits are created. For each train-test split, target is modelled as a polynomial regression function from degree 1 to 10. Each model is evaluated on test-data(complement of training data)


```python
for randomstate in range(1,11):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4,random_state=randomstate)
    for degree in range(1,max_degree+1):
        polynomial_features= PolynomialFeatures(degree=degree) ##-- degree means the highest exponent that can be possible in the polynomial
        x_poly_train = polynomial_features.fit_transform(X_train)
        x_poly_test = polynomial_features.fit_transform(X_test)
        model = LinearRegression()
        model.fit(x_poly_train, y_train)
        y_poly_pred = model.predict(x_poly_test)
        y_poly_pred_train = model.predict(x_poly_train)
        mse_test = mean_squared_error(y_test,y_poly_pred)
        mse_train = mean_squared_error(y_train,y_poly_pred_train)
        r2_test = r2_score(y_test,y_poly_pred)
        r2_train = r2_score(y_train,y_poly_pred_train)
        ################ create a plot scatter plot, y axis= mse, xaxis=degree
        mse_test_list.append(mse_test)
        mse_train_list.append(mse_train)
        #plt.scatter(X,y)
        #plt.plot(X_train,model.predict(x_poly_train))
    test_list_data.append(mse_test_list)
    train_list_data.append(mse_train_list)
    mse_test_list=[]
    mse_train_list=[]

     
```

```python
for mse_test_list in test_list_data:
    x=list(range(1,11,1))
    ax=sns.lineplot(y=mse_test_list,x=x,marker="o")
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
    ax.lines[0].set_linestyle("-")
    ax.set_xlabel('Degree of Polynomial');ax.set_ylabel('Mean Squared Error'); 
    ax.set_title("The Validation set approach-10 different random train-test splits")
    ax.set_ylim(15,32)
plt.savefig("cv_2.png",dpi=300)
```

# LOOCV 

```python
mse_test_list=[]
for degree in range(1,max_degree+1):
    polynomial_features= PolynomialFeatures(degree=degree) ##-- degree means the highest exponent that can be possible in the polynomial
    x_poly = polynomial_features.fit_transform(X)
    model = LinearRegression()
    scores = cross_val_score(model, x_poly, y, cv=X.shape[0],scoring='neg_mean_squared_error') ## cv = number of records, in case of LOOCV. LOOCV is special case of K-fold
    mse_test = abs(scores.mean())
    mse_test_list.append(mse_test)     
```

# K Fold Cross-Validation

We have created 9 different, 10 fold division of data available. For 1 10-fold iteration
- 9 parts of data are assigned as training data
- 1 part of data is assigned as testing data
- test error is evaluated on the 1-part
- this process is repeated such that each part is used to evaluate test error
- the net test error is the average of all the test errors

The following process is explained for you to understand the intution behind the CV.
for single 10 fold CV, [cross_val_score](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) can be used. 


```python
# scikit-learn k-fold cross-validation
from numpy import array
from sklearn.model_selection import KFold
```

```python
X = auto_data['horsepower'].values.reshape(-1,1) ## as the X has only one variable, it is to be reshaped
y = auto_data['mpg']
```

```python
kfold = KFold(10,True,random_state=1) ## 10 fold cv
```

```python
test_list_data=[]
mse_test_list=[]
mse_train_list=[]
## Random split into k folds 9 times
for randomstate in range(1,10):
    ## Random split into 10 folds
    kfold = KFold(10,True,random_state=randomstate)
    ## For each random split get test error
    fold_test_error_list=[]
    for train_index, test_index in kfold.split(X):
        X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
        ### for different degree of polynomial
        degree_test_error_list=[]
        for degree in range(1,max_degree+1):
            polynomial_features= PolynomialFeatures(degree=degree) ##-- degree means the highest exponent that can be possible in the polynomial
            x_poly_train = polynomial_features.fit_transform(X_train)
            x_poly_test = polynomial_features.fit_transform(X_test)
            model = LinearRegression()
            model.fit(x_poly_train, y_train)
            y_poly_pred = model.predict(x_poly_test)
            y_poly_pred_train = model.predict(x_poly_train)
            mse_test = mean_squared_error(y_test,y_poly_pred)
            mse_train = mean_squared_error(y_train,y_poly_pred_train)
            degree_test_error_list.append(mse_test)
           
        fold_test_error_list.append(degree_test_error_list)
    
    
    test_list_data.append(np.mean(fold_test_error_list,axis=0))
```

```python
test_list_data[0]
```

```python
for mse_test_list in test_list_data:
    x=list(range(1,11,1))
    ax=sns.lineplot(y=mse_test_list,x=x,marker="o")
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
    ax.lines[0].set_linestyle("-")
    ax.set_xlabel('Degree of Polynomial');ax.set_ylabel('Mean Squared Error'); 
    ax.set_title("9 different 10 Fold Cross Validation")
    ax.set_ylim(14,32)
plt.savefig("10foldcv.png",dpi=300)
```
