```python
## Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math
%matplotlib inline
```

```python
## set default style for plots
context= "notebook"
style= "darkgrid"
font_family = "Comic Sans MS"
## customizing rcParams
rcparam_dict={"figure.figsize":"25,10","axes.labelsize":1,"axes.titlesize":"large","font.family":"Comic Sans MS","text.color":"#A53010","axes.labelcolor":"A53010",
              "axes.labelsize":"medium",'xtick.color':'#A53010', 'ytick.color':'#A53010'}
sns.set(context=context,style=style,font=font_family,rc=rcparam_dict) ## Setiing context, style and rcparams in a single line
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
df_india_census= pd.read_csv("india-districts-census-2011.csv")
```

```python
#df_india_census.head().T
```

What we are concerned about is 

- Total agricultural workers in a state
- Total households
- Total Households with mobile phone

```python
df_mobile_agri_households = df_india_census.groupby("State name")["Households","Households_with_Telephone_Mobile_Phone_Mobile_only","Agricultural_Workers","Workers"].sum().reset_index()
```

```python
df_mobile_agri_households["Percentage_of_Households_with_Mobile"]=round((df_mobile_agri_households["Households_with_Telephone_Mobile_Phone_Mobile_only"]/df_mobile_agri_households["Households"])*100,2)
```

```python
df_mobile_agri_households["Percentage_of_Agri_Workers"]= round((df_mobile_agri_households["Agricultural_Workers"]/df_mobile_agri_households["Workers"])*100,2)
```

```python
df_mobile_agri_households
```

- Total Number of households vs Households with mobile
- percentage households with mobile vs agricultural workers in the state

- first plot distribution of houselholds with mobile  across states

- Then plot distribution of agricultural workers across states


```python
ax = sns.barplot(x="Percentage_of_Households_with_Mobile", y="State name", data=df_mobile_agri_households,color="#A53010")
ax.set_title("Percentage of Households with mobile phone")
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
plt.savefig("mobile_dist.png",dpi=300)
```

```python
ax = sns.scatterplot(x="Percentage_of_Households_with_Mobile", y="State name", data=df_mobile_agri_households)
ax.set_title("Percentage of Households with mobile phone")
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
```

```python
ax = sns.barplot(x="Percentage_of_Agri_Workers", y="State name", data=df_mobile_agri_households,color="#A53010")
ax.set_title("Percentage of agriculture workers")
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
plt.savefig("agri_dist.png",dpi=300)
```

```python
df_bar_plot=df_mobile_agri_households[["State name","Percentage_of_Households_with_Mobile","Percentage_of_Agri_Workers"]]
```

```python
dfs1 = pd.melt(df_bar_plot, id_vars = "State name")
```

```python
dfs1
```

```python
ax=sns.barplot(x = 'value', y='State name', hue = 'variable',data=dfs1)
ax.set_title("Percentage of households with mobile Vs Percentage of Agricultural workers in Indian States/Uts")
ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
plt.legend(bbox_to_anchor=(0.88, 1), loc=2, borderaxespad=0.)
plt.savefig("mobile_agri_dist.png",dpi=300)
```
