## Adding Axes and Title to the plot
- plot titile
- x axis label
- y axis label

```python
sns.set_style("darkgrid")
ax = sns.scatterplot(x="total_bill", y="tip", data=data)
ax.set_xlabel("Total Bill (USD)")
ax.set_ylabel("Tip (USD)")
ax.set_title("Tip Vs Total Bill")
```

## This Plot look a little better. But it is small. How can I use it for a Poster? or for a Paper?
### What is RcParams?
Through rcParams function, we can overwrite the default values of underlying matplotlib parameters that control the look and feel of the plot.
### Here are some of the important properties which can be controlled through rcParams are
- font.family
- text.color
- axes.labelcolor
- xtick.color
- ytick.color
- font.size
- figure.figsize
- axes.labelsize
- axes.titlesize

## The best way is to google the rcParams as per your need and set it as I will do in the following code snippet

```python
## customizing rcParams
rcparam_dict={"figure.figsize":"16,10","axes.labelsize":50,"axes.titlesize":"large","font.family":"Comic Sans MS","text.color":"#A53010","axes.labelcolor":"A53010",
              "axes.labelsize":"medium",'xtick.color':'#A53010', 'ytick.color':'#A53010'}
```

## What is Context in Seaborn?
There are 4 contexts available in seaborn which may be used as per need.This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style. The base context is 
- “notebook”, and the other contexts are 
- “paper”, 
- “talk”, and 
- “poster”, which are version of the notebook parameters scaled by .8, 1.3, and 1.6, respectively. 

## Set() function in seaborn
Detailed documentation of set() funcation can be found in the following link -
- https://seaborn.pydata.org/generated/seaborn.set.html

We can set context, style, font and rcParams all in one line using seaborn set function
Let's do it

```python
## defining
context= "poster"
style= "darkgrid"
font_family = "Comic Sans MS"
```

```python

sns.set(context=context,style=style,font=font_family,rc=rcparam_dict) ## Setiing context, style and rcparams in a single line
ax = sns.scatterplot(x="total_bill", y="tip", data=data,facecolor="blue")
ax.set_xlabel("Total Bill (USD)")
ax.set_ylabel("Tip (USD)")
ax.set_title("Tip Vs Total Bill")
```

## How to control x and y axes gridline or ticks spacing?

```python
start_x,end_x = ax.get_xlim() # get x axis minimum and maximum values as per data
```

```python
start_x
```

```python
start_y,end_y = ax.get_ylim() ## get y axis minimum and maximum values as per data
```

```python
stepsize_x= 5 # Let's say we want to increment x axes ticks by 5
stepsize_y= 1 # Let's say we want to increment x axes ticks by 5
```

```python
x_ticks=np.arange(round(start_x) ,round(end_x)+1 ,stepsize_x)
y_ticks= np.arange(round(start_y) ,round(end_y)+1 ,stepsize_y)
```

```python
x_ticks
```

```python
y_ticks
```

```python
sns.set(context=context,style=style,font=font_family,rc=rcparam_dict) ## Setiing context, style and rcparams in a single line
ax = sns.scatterplot(x="total_bill", y="tip", data=data,hue="sex",palette=["red","blue"])
ax.set_xlabel("Total Bill (USD)")
ax.set_ylabel("Tip (USD)")
ax.set_title("Tip Vs Total Bill")
ax.xaxis.set_ticks(x_ticks)
ax.yaxis.set_ticks(y_ticks)
sns.despine()
```
